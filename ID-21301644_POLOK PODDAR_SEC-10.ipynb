```python
#Question_3
def group_up(x):
    a = []
    if not x:
        return a
    nums = [''.join(sorted(word)) for word in x]

    d = {}
    for i, e in enumerate (nums):
        d.setdefault(e, []).append(i)
    print(d) 
    for j in d.values():
        collection = tuple(x[i] for i in j)
        if len(collection) > 1:
            a.append(collection)
    return a
words =  ["ant", "ate", "ozo", "eat", "tan", "tea", "tax", "zoo"]
a = group_up(words)
pro=[]
for y in a:
    y=list(y)
    pro.append(y)
print(pro)
```
